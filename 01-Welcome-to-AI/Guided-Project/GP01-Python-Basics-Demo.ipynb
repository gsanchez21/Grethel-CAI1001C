## Part 2: Basic Math with Python

Python can do math like a calculator.

```python
# Basic arithmetic
price = 15.99
quantity = 3

# Calculate total
total = price * quantity
print(f"Price per item: ${price}")
print(f"Quantity: {quantity}")
print(f"Total: ${total}\n")

# Add sales tax (7%)
tax_rate = 0.07
tax = total * tax_rate
final_total = total + tax

print(f"Tax (7%): ${tax:.2f}")
print(f"Final Total: ${final_total:.2f}")
```

### 📝 What's Happening?

- `*` multiplies numbers
- `+` adds numbers
- `.2f` in the f-string formats numbers to 2 decimal places (for money)

```python
# Calculate average temperature for the week
monday = 65
tuesday = 65
wednesday = 73
thursday = 78
friday = 83

# Add them up
total_temp = monday + tuesday + wednesday + thursday + friday

# Divide by number of days
average_temp = total_temp / 5

print(f"Total temperature: {total_temp}°F")
print(f"\nAverage temperature: {average_temp}°F")
```

## Part 3: Lists - Storing Multiple Values

Lists hold multiple items in order.

```python
# Create a list of temperatures
temperatures = [75, 82, 79, 85, 88]

print("This week's temperatures:")
print(temperatures)

# Access individual items (Python counts from 0!)
print(f"\nFirst day: {temperatures[0]}°F")
print(f"Second day: {temperatures[1]}°F")
print(f"Last day: {temperatures[4]}°F")
```

### 📝 What's Happening?

- Square brackets `[]` create a list
- Lists can hold multiple values
- Access items with `list_name[index]`
- **Python counts from 0**, so the first item is index 0!

```python
# List of student names
students = ["Carlos", "Maria", "Sofia", "Roberto", "Ana"]

print("Students in class:")
print(students)

print(f"\nTotal students: {len(students)}")
print(f"First student: {students[0]}")
print(f"Last student: {students[-1]}")  # -1 means last item!
```

### 💡 Cool Trick

- `len()` tells you how many items in a list
- Negative indices count backwards: `[-1]` is the last item!

## Part 4: Analyzing Lists

Python has built-in functions to analyze lists of numbers.

```python
# Quiz scores
scores = [85, 92, 78, 95, 88]

print("Quiz Scores:")
print(scores)

# Built-in functions for analysis
total = sum(scores)          # Add all scores
average = total / len(scores) # Calculate average
highest = max(scores)         # Find highest score
lowest = min(scores)          # Find lowest score

print(f"\nTotal points: {total}")
print(f"Average score: {average:.1f}")
print(f"Highest score: {highest}")
print(f"Lowest score: {lowest}")
```

### 📝 What's Happening?

- `sum()` adds all numbers in a list
- `len()` counts items in a list
- `max()` finds the largest number
- `min()` finds the smallest number

These are the building blocks of data analysis!

```python
# Monthly expenses (in dollars)
expenses = [2143, 800, 525, 510, 200, 200]

# Calculate monthly budget info
total_expenses = sum(expenses)
average_expense = total_expenses / len(expenses)
highest_expense = max(expenses)

print("Monthly Expense Analysis:")
print(f"Total expenses: ${total_expenses:.2f}")
print(f"Average per category: ${average_expense:.2f}")
print(f"Highest expense: ${highest_expense:.2f}")
```

## Part 5: For Loops - Processing Lists

Loops let you do something with each item in a list.

```python
# List of daily temperatures
temperatures = [75, 82, 79, 85, 88]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

print("This Week's Weather:")
print()

# Loop through both lists at the same time
for list in range(len(days)):
    day = days[list]
    temp = temperatures[list]
    print(f"{day}: {temp}°F")
```

### 📝 What's Happening?

- `for i in range(len(days)):` creates a loop that runs 5 times (0, 1, 2, 3, 4)
- Each time through, `i` changes (0, then 1, then 2, etc.)
- We use `i` to get the item from each list
- The indented code runs each time through the loop

```python
# Simpler way - loop directly through a list
fruits = ["mango", "papaya", "guava", "maracuyá"]

print("Fruits I love:")
for fruit in fruits:
    print(f"  - {fruit}")
```

### 📝 What's Happening?

- `for fruit in fruits:` loops through each item
- Each time through, `fruit` gets the next item
- First time: `fruit = "mango"`
- Second time: `fruit = "papaya"`
- And so on...

```python
# Calculate and print individual percentages
scores = [85, 92, 78, 95, 88]
total_possible = 100

print("Quiz Percentages:")
for score in scores:
    percentage = (score / total_possible) * 100

    # Determine the label based on the score
    if percentage >= 90:
        label = "Excellent"
    elif percentage >= 80:
        label = "Good"
    elif percentage >= 70:
        label = "Passing"
    else:
        label = "Failing"

    print(f"Score {score} = {percentage:.1f}% - {label}")
```

## Part 6: Combining Everything - A Complete Example

Let's use variables, lists, loops, and calculations together!

```python
# Miami Beach Weekly Temperature Analysis

# Data for one week
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [75, 82, 79, 85, 88, 91, 87]

print("="*50)
print("MIAMI BEACH WEEKLY TEMPERATURE REPORT")
print("="*50)

# Display each day
print("\nDaily Temperatures:")
for i in range(len(days)):
    print(f"  {days[i]}: {temperatures[i]}°F")

# Calculate statistics
total_temp = sum(temperatures)
average_temp = total_temp / len(temperatures)
highest_temp = max(temperatures)
lowest_temp = min(temperatures)

print("\n" + "-"*50)
print("Weekly Statistics:")
print("-"*50)
print(f"Average temperature: {average_temp:.1f}°F")
print(f"Highest temperature: {highest_temp}°F")
print(f"Lowest temperature: {lowest_temp}°F")
print(f"Temperature range: {highest_temp - lowest_temp}°F")

# Count hot days (85°F or higher)
hot_days = 0
for temp in temperatures:
    if temp >= 85:
        hot_days += 1

print(f"\nHot days (85°F+): {hot_days} out of {len(days)}")
print("="*50)
```

### 📝 What's Happening?

This example combines:
1. **Two lists** - days and temperatures
2. **A loop** - to display each day
3. **Calculations** - sum, average, max, min
4. **Another loop with a counter** - counting hot days
5. **Formatted output** - making it look professional

This is the kind of data analysis you'll do with AI!

## Count Expensive Days

### Subtask:
Iterate through your `spending` list to count and display the number of days where your spending exceeded $40.


**Reasoning**:
To fulfill the subtask, I will write Python code that initializes a counter, iterates through the 'spending' list, increments the counter for each day where spending exceeds $40, and then prints the final count.



```python
expensive_days_count = 0

print("\n--- Analyzing Expensive Days ---")
for day_spending in spending:
    if day_spending > 40:
        expensive_days_count += 1

print(f"Number of days with spending over $40: {expensive_days_count}")
```

## Analyze Weekday vs. Weekend Spending

### Subtask:
Separate your `spending` data into weekday and weekend categories, then calculate and display the average spending for each period to compare them.


**Reasoning**:
To fulfill the subtask, I will write Python code that separates the `spending` data into `weekday_spending` and `weekend_spending` lists based on the `days` list, calculates the total and average spending for each period, and then prints these averages.



```python
weekday_spending = []
weekend_spending = []

print("\n--- Analyzing Weekday vs. Weekend Spending ---")
for i in range(len(days)):
    if days[i] in ["Saturday", "Sunday"]:
        weekend_spending.append(spending[i])
    else:
        weekday_spending.append(spending[i])

# Calculate totals
total_weekday_spending = sum(weekday_spending)
total_weekend_spending = sum(weekend_spending)

# Calculate averages
# Avoid division by zero if there are no weekdays/weekends in the data
average_weekday_spending = total_weekday_spending / len(weekday_spending) if len(weekday_spending) > 0 else 0
average_weekend_spending = total_weekend_spending / len(weekend_spending) if len(weekend_spending) > 0 else 0

print(f"Average weekday spending: ${average_weekday_spending:.2f}")
print(f"Average weekend spending: ${average_weekend_spending:.2f}")
```

## Perform Budget Check

### Subtask:
Compare your `total_spending` with a weekly budget of $200 and print whether you are over or under the budget, along with the difference.


**Reasoning**:
To fulfill the subtask, I will define a weekly budget, calculate the difference with the total spending, and then use an if-else statement to compare the total spending with the budget and print the appropriate message.



```python
weekly_budget = 200
budget_difference = total_spending - weekly_budget

print("\n--- Budget Check ---")
if total_spending > weekly_budget:
    print(f"You are over budget by ${budget_difference:.2f}")
else:
    print(f"You are under budget by ${abs(budget_difference):.2f}")
```
